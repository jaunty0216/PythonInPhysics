```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
pp = np.pi
```

```python
X = np.linspace(-pp,pp,256)
line1 = 0.9*X
line2 = 1.1*X
line3 = np.tanh(X)
plt.plot(X,line1,color="black",linestyle = "-")
plt.plot(X,line2,color="black",linestyle = "--")
plt.plot(X,line3,color="black",linestyle = ":")
plt.show()
```

```python
n = 256
pp = np.pi
x = np.linspace(-5, 5, n)
z = np.linspace(-5, 5, n)
y = 0
d0 = 1
q = 4
eps0 = 1/pp#定義會用到的東西

X, Z = np.meshgrid(x, z)
#重點：contour
def calculating(x,z,q,eps,d):
    first = np.sqrt(x**2 + (z - d/2)**2)
    second = np.sqrt(x**2 + (z + d/2)**2)
    V = (q/4*pp*eps)*(1/first - 1/second)
    return V

myfunction = calculating(X,Z,q,eps0,d0)
contour_levels = [-0.2,-0.15,-0.1,-0.05,0,0.05,0.1,0.15,0.2]
plt.contour(X, Z, myfunction, levels=contour_levels, colors='blue')
```

```python
n = 30
pp = np.pi
x = np.linspace(-5, 5, n)
z = np.linspace(-5, 5, n)
y = 0
d0 = 1
q = 4
eps0 = 1/pp

X, Z = np.meshgrid(x, z)
#重點：contour
def calculating(x,z,q,eps,d):
    
    first = np.sqrt(x**2 + (z - d/2)**2)
    second = np.sqrt(x**2 + (z + d/2)**2)
    #計算|r-r0|
    Ex = (q/4*pp*eps)*(x/first**3 - x/second**3)
    Ez = (q/4*pp*eps)*((z - d/2)/first**3 - (z + d/2)/second**3)
    #計算r-r0
    return Ex,Ez

Ex,Ez = calculating(X,Z,q,eps0,d0)
plt.quiver(X,Z, Ex, Ez)
```

```python
fig = plt.figure()  # 創建一個新的圖形
ax = fig.add_subplot(111, projection='3d')  # 在圖形中添加一個3D子圖
#建立3D物件
Ps = np.linspace(0,2*pp,256)
Ph = np.linspace(0,2*pp,256)
R = 2
a = 1
xx = np.linspace(-5,5,1)
yx = np.linspace(-5,5,1)
ps,ph = np.meshgrid(Ps,Ph)
#初始化
x = (R + a*np.sin(ps))*np.cos(ph)
y = (R + a*np.sin(ps))*np.sin(ph)
z = a*np.cos(ps)
#the環
ax.plot_surface(x,y,z,cmap='viridis')#重點2，畫出來
plt.show()  # 顯示圖形
```

```python
fig = plt.figure()  # 創建一個新的圖形
ax = fig.add_subplot(111, projection='3d')  # 在圖形中添加一個3D子圖
#建立3D物件
Ps = np.linspace(0,2*pp,256)
Ph = np.linspace(0,2*pp,256)
R = 3
a = 1
x = np.linspace(-5,5,1)
y = np.linspace(-5,5,1)
ps,ph = np.meshgrid(Ps,Ph)
#初始化
x = (R + a*np.sin(ps))*np.cos(ph)
y = (R + a*np.sin(ps))*np.sin(ph)
z = a*np.cos(ps)
#the環
result = ax.plot_surface(x,y,z,cmap='viridis')#重點2，畫出來
fig.colorbar(result, shrink=0.5, aspect=5)
# 設定視角
ax.view_init(elev=55, azim=30)#elev是仰角，azim是方位角。
plt.show()  # 顯示圖形
```
