```python
%matplotlib inline 
#引入matplotlib
import numpy as np
import matplotlib.pyplot as plt #matplotlib有很多東西，拿取其中的pyplot
#前面是起手式
X = np.linspace(-np.pi, np.pi, 256, endpoint=True) #定義定義域：一個範圍，在256個點之間(切成255份)，從-pi到pi 重點：arange
C, S = np.cos(X), np.sin(X) #分別定義C和S是甚麼
plt.plot(X, C) #進行描繪作業
plt.plot(X, S) #重點： .plot(範圍,函數)
plt.show() #展現出來
```

```python
# Fixing random state for reproducibility
np.random.seed(196802)
N = 100
r0 = 0.6
x = 0.9 * np.random.rand(N) #0 < x < 0.9
y = 0.9 * np.random.rand(N) #0 < x < 0.9
area = (20 * np.random.rand(N))**2 # 0 < area < 400
c = np.sqrt(area) # 0 < c < 20
r = np.sqrt(x ** 2 + y ** 2)#0 < r < 1.27，或點和原點的距離

area1 = np.ma.masked_where(r < r0, area)#符合條件將被遮罩
area2 = np.ma.masked_where(r >= r0, area)
#假設r=0.3，在area1，0.3<r0，True，所以會被遮罩。到了area2，0.3>=r0，False，所以會被保留，所以屬於area2
#假設r=0.7，在area1，0.7<r0，False，所以會被保留。到了area2，0.7>=r0，True，所以會被遮罩，所以屬於area1
#這兩行結論：把小於0.6的部分歸類在area2；大於0.6的部分歸類在area1。
#area1或area2其實都還是area，只是範圍變了
plt.scatter(x, y, s=area1, marker='^', c=c)
plt.scatter(x, y, s=area2, marker='o', c=c)

# Show the boundary between the regions:
theta = np.arange(0, np.pi / 2, 0.01)
plt.plot(r0 * np.cos(theta), r0 * np.sin(theta))#(0.6~0,0~0.6)
#一個四分之一圓，表示r0的範圍
plt.show()
```

```python
n = 12
X = np.arange(n) #X=0,1,2...,11
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)#Y1和Y2其實都是陣列，Y1 = 1,11/12,10/12...1/12
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)#uniform的意思就是產生隨機亂數，和rand不同的是可以自訂產生範圍

plt.bar(X, Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='black')#本期重點

for x, y in zip(X, Y1):#zip的功能：把兩個陣列合併在一起。如果合併的內容一長一短，則只會輸出到短的量。
    plt.text(x + 0.4, y + 0.05, '%.2f ' % y, ha='center',
    va='bottom')
    #重點：'%.2f ' % y，控制y輸出到小數點後兩位
    #ha = horizontal alignment，水平對齊
    #va = vertical alignment，垂直對齊
for x, y in zip(X, -Y2):#zip的功能：把兩個陣列合併在一起。如果合併的內容一長一短，則只會輸出到短的量。
    plt.text(x , y-0.25 , '%.2f ' % y, ha='center',
    va='bottom')
plt.ylim(-1.25, 1.5)
```

```python
def f(x, y):#設f(x,y)...
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)
    
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)#把-3到3之間切成256等分

X, Y = np.meshgrid(x, y)#生成二維座標網格，會回傳兩個二維數組，分別指派給X和Y
plt.contourf(X, Y, f(X, Y),7 , alpha=0.75, cmap='jet')#畫出的是等值區域圖，f應該是指"field"
plt.contour(X, Y, f(X, Y), 7, colors='black')#畫出的是等高線
```

```python
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)#-10到9
U, V = np.meshgrid(X,Y)

fig,ax = plt.subplots()#創建圖形用的命令
plt.figure(figsize=(10,6),dpi = 80)

ax.set_title('Arrows scale with plot width')#設定標題
q = ax.quiver(X,Y,-V,U)#產生箭頭
ax.quiverkey(q, 0.85, 1.05, 10,r'$2 (m/s)$', labelpos='E')#labelpos裡的英文是方位，這裡意思是"在箭頭的東側"。
#fig.savefig('testinggggg.png') #這命令會保存照片，放在現在這份程式碼文件所在的母資料夾
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  # 導入3D軸類
fig = plt.figure()  # 創建一個新的圖形
ax = fig.add_subplot(111, projection='3d')  # 在圖形中添加一個3D子圖

r = np.linspace(0, 1.25, 50)  # 創建一個包含50個從0到1.25的數字的等差數列
p = np.linspace(0, 2*np.pi, 50)  # 創建一個包含50個從0到2π的數字的等差數列

R, P = np.meshgrid(r, p)  # 生成網格點座標矩陣
Z = ((R**2 - 1)**2)  # 定義Z軸上的函數
X, Y = R*np.cos(P), R*np.sin(P)  # 將極座標轉換為笛卡爾座標

ax.plot_surface(X, Y, Z, cmap=plt.cm.YlOrRd_r)  # 繪製3D曲面圖，並使用顏色映射
#關於plt.cm.YlGnBu_r：黃經綠到藍，反轉

ax.set_zlim(0,1)  # 設定Z軸的限制

ax.set_xlabel(r'$\phi\mathrm{real}$')# 設定X軸標籤
ax.set_ylabel(r'$\phi_\mathrm{im}$')# 設定Y軸標籤
ax.set_zlabel(r'$V(\phi)$')# 設定Z軸標籤
ax.view_init
plt.show()  # 顯示圖形
```
